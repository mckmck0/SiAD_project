# TEST STATYSTYCZNY


Hipoteza zerowa H0:
Marki z pierwszej 5 najczęściej sprzedawanych produktów nie mają żadnego wpływu na stosunek ilości
sprzedaży do wyświetleń produktu. (H0: p = p0)
Hipoteza alternatywna H1:
Marki (product_brand) z pierwszej 5 najczęściej sprzedanych produktów znacznie różnią się stosunkiem
ilości sprzedaży do wyświetleń produktu.    (H1: p != p0)
Poziom istotności Alfa = 0.05.


```python
df_exp = df_exp[df_exp['product_brand'] != '-1']
df_exp_not_sold = df_exp[df_exp['Sale'] == 0]
```

```python
# df_sub = df_exp.groupby(['product_brand', 'Sale']).size()
df_exp.dropna(axis='rows')
df_sub = pd.crosstab(df_exp['product_brand'], np.where(df_exp['Sale'] == 0, 'View_count', 'Sale_count')).reset_index()
df_sub['Sale_View_ratio'] = df_sub['Sale_count'] / (df_sub['View_count'] + df_sub['Sale_count'])
df_sub.replace(np.inf, 0, inplace=True)

df_sub.describe()
```

Wykonamy teraz test z dla proporcji.

```python
def z_test(sample_proportion, h0_proportion, sample_size):
    z = (sample_proportion - h0_proportion) \
        / np.sqrt((h0_proportion * (1 - h0_proportion)) / sample_size)

    return z
```

```python
z_score = z_test(top_5_sale_view_ratio, sale_view_ratio,
                          top_5_view_count + top_5_sale_count)
print("Wartość statystyki testowej Z: {}\n".
      format(round(z_score, 3)))
```

# PCA
Po stworzeniu nowych cech datasetu, czas na przesianie szumu - nieistotnych zmiennych.
Użyję do tego dekompozycji widmowej w implementacji biblioteki sklearn


Jednak przed użyciem PCA, warto usunąć silnie skorelowane kolumny.

```python
df_pca = feature_matrix
# df_pca
corr=df_pca.corr()
corr
```

```python
threshold = 0.8
triangle = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
collinears = [c for c in triangle.columns if any(abs(triangle[c]) > threshold)]
collinears
df_pca = df_pca.drop(df_pca[collinears], axis=1)
```

Jako, że przygotowujemy się do zredukowania wymiarowości datasetu za pomocą PCA, musimy usunąć z niego

Jako, że PCA działa najlepiej na danych znormalizowanych, przeprowadźmy ją za pomocą StandardScalera
w implementacji sklearn.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca = PCA()
scaler = StandardScaler()
X_b = scaler.fit_transform(X)
# print(X)

pca.fit_transform(X_b)
```

Po pomyślnym wykonaniu dekompozycji, sprawdźmy ile głównych cech powinniśmy wybrać
dla dalszej analizy.

```python
feats = range(pca.n_components_)
var_sum = np.cumsum(pca.explained_variance_ratio_)
print(var_sum)
```

Teraz zobaczmy to na wykresie

```python
plt.plot(var_sum)
plt.axhline(y=0.9, color='r')
plt.xlabel('Number of features')
plt.ylabel('Cumulative variance')
plt.show()
```

Na powyższym wykresie możemy zauważyć, że pierwsze ok. 40 cech zawiera 90% informacji o datasecie.
Wybieram zatem tę liczbę jako liczbę komponentów do dalszej analizy.

Teraz trzeba podzielić nasze dane na dane testowe i treningowe. Użyję do tego funkcji train_test_split
z biblioteki sklearn. Model będziemy trenować na 70% danych.
Dane treningowe trzeba ponownie znormalizować.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=420)
scaler = StandardScaler()
scaler.fit(X_train)

scaler.transform(X_train)
scaler.transform(X_test) # na danych testowych przeprowadzamy tylko transformację,
                         # żeby nasz model nie znał parametrów danych testowych
pca_f = PCA(n_components=40)
pca_f.fit(X_train)
```

```python
print(pca_f.explained_variance_,
      "\n-------------------------------------------------------------------\n",
      pca_f.explained_variance_ratio_)
```

```python
X_train = pca_f.transform(X_train)
X_test = pca_f.transform(X_test)
```

Naszym problemem jest przyporządkowanie czy dany produkt został sprzedany - czy nie.
Jest to zatem problem klasyfikacji.
Najpierw, żeby mieć odnośnik do porównania wyników, użyję prostej regresji logistycznej
w implementacji sklearn.
Potem porównam wyniki z biblioteką TPOT.

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=420)
lr.fit(X_train, Y_train)
Y_pred = lr.predict(X_test)
```

Sprawdźmy wynik predykcji.

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error
score_lr = lr.score(X_test, Y_test)
print("Model accuracy: {}".format(round(score_lr, 3)))
y_test = [int(x) for x in Y_test]
y_pred = [int(x) for x in Y_pred]
rmse = np.sqrt(mse(y_test, y_pred))
print("Prediction RMSE: {}".format(round(rmse, 3)))
```

Pomimo, że model nie był w żaden sposób tuningowany, udało sie osiągnąć celność prawie 90%
i niską wartość średniej kwadratowej błędów (RMSE).

TPOT szuka najlepszego modelu poprzez sprawdzian krzyżowy (Cross Validation) i na końcu zwraca
najcelniejszy model (cały pipeline modelu wraz ze wszystkimi przetestowanymi hiperzmiennymi).

Po 10 minutach TPOT za najskuteczniejszy model uznał 3 razy uznał DecisionTreeClassifier z celnościami: 99.7%, 95.5%, 90.3%.
Najskuteczniejsza okazała się ostatnia próba z wybranym modelem ExtraTreesClassifier.
